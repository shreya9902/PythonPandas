```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando","baltimore"],
    "temperature": [21,14,35,45],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando","sans fransisco"],
    "humidity": [65,68,75,56],
})
df2
```

# Merge

```python
df=pd.merge(df1,df2, on="city")
df
```

```python
df=pd.merge(df1,df2, on="city", how="outer")
df
```

```python
df=pd.merge(df1,df2, on="city", how="inner")
df
```

```python
df=pd.merge(df1,df2, on="city", how="left")
df
```

```python
df=pd.merge(df1,df2, on="city", how="right")
df
```

```python
df=pd.merge(df1,df2, on="city", how="outer", indicator=True)
df
```

```python
df3= pd.merge(df1,df2, on="city")
df3
```

# Join

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1.set_index('city',inplace=True)
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2.set_index('city',inplace=True)
df2
```

```python
df1.join(df2,lsuffix='_l', rsuffix='_r')
```
